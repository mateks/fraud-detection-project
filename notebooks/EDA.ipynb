# Download raw data

```python
import kagglehub

path = kagglehub.dataset_download("sriharshaeedala/financial-fraud-detection-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np

dataset = pd.read_csv(path + "/Synthetic_Financial_datasets_log.csv")
```

```python
print(dataset.info())
print(dataset.describe())
print(dataset.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.displot(x="step", col="isFraud", data = dataset, kind="kde", fill=True, height=4, aspect=1.5)
plt.show()
```
